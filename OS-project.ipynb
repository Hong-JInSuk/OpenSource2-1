```python
testx_r=np.array(exam.iloc[:,[6]])
testy_r=exam['reading score']
testx_w=np.array(exam.iloc[:,[7]])
testy_w=exam['writing score']
testx_m=np.array(exam.iloc[:,[5]])
testy_m=exam['math score']

train_df=exam
x_w=np.array(train_df.iloc[:,[7]])
y_w=train_df['writing score']
x_r=np.array(train_df.iloc[:,[6]])
y_r=train_df['reading score']
x_m=np.array(train_df.iloc[:,[5]])
y_m=train_df['math score']
```

```python
edu_w=np.median(y_w)

class_yw = [1 if i > edu_w else 0 for i in y_w]
#print(class_y)

test_class_yw= [1 if i > edu_w else 0 for i in testy_w]
#print(test_class_y)
```

```python
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression(random_state=0).fit(x_m, class_yw)
```

```python
predicted_y_mw=clf.predict(testx_m)
#print(predicted_y_mw)
predicted_prob_y_mw=clf.predict_proba(testx_m)
#print(predicted_prob_yw)
```

```python
clf.score(testx_m, test_class_yw)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(test_class_yw, predicted_y_mw)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_class_yw, predicted_y_mw))
```

```python
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc=roc_auc_score(test_class_yw, predicted_yw)
fpr,tpr,thresholds=roc_curve(test_class_yw, predicted_prob_y_mw[:,1])
print(logit_roc_auc)
plt.figure(figsize=(5,5))
plt.plot(fpr,tpr,label='Logistic Regression (area = %0.2f)'%logit_roc_auc)
plt.plot([0,1],[0,1],'r--')
plt.xlabel('< FPR-Math >')
plt.ylabel('< TPR-Writing >')
plt.title('Logistic')
```
